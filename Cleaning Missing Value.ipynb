```python
df = pd.read_csv('plays.csv', error_bad_lines=False)
df.head(3)
```

```python
df_backup = df.copy(deep=True)
#df.copy fungsinya untuk membuat salinan data frame di memori tidak di ram
```

```python
df_backup.isna().mean()
#.isna() untuk menampilkan missing values dan .mean() untuk merata-rata kan jumlah missing value 
```

```python
def missing_percentage():
    na_val = df_backup.isnull().sum() * 100 / len(df_backup)
    na_val = na_val[na_val.values > 0]
    na_val = na_val.sort_values(ascending=False)
    return na_val
```

```python
missing_percentage()
```

```python
df_backup = df.drop(columns=['city', 'utm_campaign', 'utm_source', 'utm_medium', 'genre_name', 'film_title',
                             'season_name', 'hash_film_id', 'app_version', 'stream_type',])
```
